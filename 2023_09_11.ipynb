### 데이터프레임 확장
- concat : 그냥 가져다 붙이는 거 　# 행, 열을 추가할 때
- merge : 공통된 칼럼이나 인덱스가 있는 경우

```python
s1 = pd.Series([500000, 500500, 501000], index = dates, name = 'LG화학')
```

```python
concat_df2 = pd.concat([df2, s1], axis = 1)   # axis = 0 행, axis = 1 열방향으로 동작
concat_df2
```

```python
concat_df2['LG화학_add_500'] = concat_df2['LG화학'] + 500
concat_df2
```

```python
concat_df2.loc['2023-01-06',:] = [44000, 70100, 85000, 455000, 505000, 501000]
concat_df2
```

```python
concat_df2.loc['2023-01-05',:] = concat_df2.loc['2023-01-06',:] - 100
concat_df2
```

### 결측치 처리
- 제거 : dropna
- 대체 : fillna, interpolate

```python
concat_df2.sort_index(inplace = True)
```

```python
import numpy as np
concat_df2.loc['2023-01-05',"LG전자":'LG화학'] = np.NaN
concat_df2
```

```python
# dropna()는 제거NaN이 포함된 행이 날라감.
concat_df2.dropna()
```

```python
# 열을 날리려면 axis를 넣기
concat_df2.dropna(axis = 1)
```

```python
# 채우기
concat_df2.fillna(method='bfill') # 'ffill' 앞의 값, 'bfill' 이전 값
```

```python
concat_df2.interpolate() #전후 값을 고려하여 중간값으로 적절히 채워주는 방법
```

### 중복 제거

```python
# 중복데이터를 생성합니다.
concat_df3 = concat_df2.copy()
concat_df4 = pd.concat([concat_df2, concat_df3])
concat_df4
```

```python
# 중복데이터 날려주기
concat_df4.drop_duplicates()
```

```python
concat_df2.loc['2023-01-03',:] = concat_df2.loc['2023-01-06',:].copy()
concat_df2.sort_index(inplace = True)
```
